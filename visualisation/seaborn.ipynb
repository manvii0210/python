```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('country_wise_latest.csv')
df.head()
```

```python
#1. Bar Plot: Total Confirmed Cases by Country
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(data=df.nlargest(10, 'Confirmed'), x='Country/Region', y='Confirmed', palette='viridis')
plt.xticks(rotation=45)
plt.title('Top 10 Countries with Most Confirmed Cases')
plt.show()
```

```python
#2. Scatter Plot: Deaths vs. Confirmed Cases
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Confirmed', y='Deaths', hue='WHO Region', size='Deaths', sizes=(10, 200), alpha=0.7)
plt.xscale('log')
plt.yscale('log')
plt.title('Deaths vs. Confirmed Cases')
plt.show()
```

```python
#3. Line Plot: New Cases Over the Last Week (for Selected Countries)
top_countries = df.nlargest(5, 'Confirmed')['Country/Region']
df_selected = df[df['Country/Region'].isin(top_countries)]

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_selected, x='Country/Region', y='New cases', marker='o', hue='Country/Region')
plt.xticks(rotation=45)
plt.title('New Cases Over Last Week')
plt.show()
```

```python
#4. Pie Chart: Active Cases Distribution Among Top 5 Countries
top5 = df.nlargest(5, 'Active')
plt.figure(figsize=(6, 6))
plt.pie(top5['Active'], labels=top5['Country/Region'], autopct='%1.1f%%', colors=sns.color_palette("Set2"))
plt.title('Active Cases Distribution Among Top 5 Countries')
plt.show()
```

```python
#5. Box Plot: Spread of New Cases Across WHO Regions
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='WHO Region', y='New cases', palette='coolwarm')
plt.xticks(rotation=45)
plt.yscale('log')
plt.title('Distribution of New Cases Across WHO Regions')
plt.show()
```

```python
#7. Stacked Bar Plot: WHO Region-wise Cases Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='WHO Region', weights='Confirmed', hue='WHO Region', multiple='stack', shrink=0.8)
plt.xticks(rotation=45)
plt.title('Confirmed Cases Distribution by WHO Region')
plt.show()
```

```python
#8. Violin Plot: Deaths per 100 Cases Across WHO Regions
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='WHO Region', y='Deaths / 100 Cases', palette='pastel')
plt.xticks(rotation=45)
plt.title('Deaths per 100 Cases Across WHO Regions')
plt.show()
```

```python
#9. Bar Plot: 1 Week % Increase by Country
plt.figure(figsize=(12, 6))
sns.barplot(data=df.nlargest(10, '1 week % increase'), x='Country/Region', y='1 week % increase', palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Top 10 Countries with Highest % Increase in Cases (Last Week)')
plt.show()
```

```python
#10. KDE Plot: Distribution of Deaths per 100 Cases
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='Deaths / 100 Cases', fill=True, color='red')
plt.title('Density Plot of Death Rate per 100 Cases')
plt.show()
```

```python
#6. Heatmap: Correlation Between Different Metrics
import numpy as np

plt.figure(figsize=(10, 6))
sns.heatmap(df[['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases', 'New deaths', '1 week change']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of COVID Metrics')
plt.show()
```
